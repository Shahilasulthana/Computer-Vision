```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("threshold.jpg", cv2.IMREAD_GRAYSCALE)
```

**1.Manual Thresholding with a threshold value of 100**

we initialize the particular threshold value and that threshold is distributed to the overall image

```python
_, manual_thresh = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
```

**2.Otsu's Thresholding**

It automatically chooses the best threshold value from the entire image and assign that threshold value to the entire image

```python
_, otsu_thresh = cv2.threshold(
    img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)
```

**3.Mean Adaptive Thresholding**

Chooses the threshold value from small local regions and distributes that threshold value to the entire image

```python
adaptive_mean = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    blockSize=11,
    C=2
)
```

**3.Gaussian adaptive thrisholding with block sizes 15, 25 and 35**

It is similar to mean adaptive thresholding but chooses the weighted average of the mean value from the nearby pixels or it gives more importance to the neighbourhood pixels

```python
gaussian_15 = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    blockSize=15,
    C=2
)

gaussian_25 = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    blockSize=25,
    C=2
)

gaussian_35 = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    blockSize=35,
    C=2
)
```

```python
titles = [
    "Original_image",
    "Manual threshold with a threshold value of 100",
    "Otsu's Thresholding",
    "Mean adaptive thresholding ",
    "Gaussian adaptive 15",
    "Gaussian adaptive 25",
    "Gaussian adaptive 35"
]

images = [
    img,
    manual_thresh,
    otsu_thresh,
    adaptive_mean,
    gaussian_15,
    gaussian_25,
    gaussian_35
]

plt.figure(figsize=(14, 8))

for i in range(len(images)):
    plt.subplot(2, 4, i + 1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Titles for histogram plots
hist_titles = [
    "Original_image",
    "Manual threshold with a threshold value of 100",
    "Otsu's Thresholding",
    "Mean adaptive thresholding ",
    "Gaussian adaptive 15",
    "Gaussian adaptive 25",
    "Gaussian adaptive 35"
]

plt.figure(figsize=(14, 10))

for i in range(len(images)):
    plt.subplot(4, 2, i + 1)
    plt.hist(images[i].ravel(), bins=256, range=[0, 256])
    plt.title(hist_titles[i])
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Gaussian adaptive thresholding is the overall best thresholding as it shows more finer details about the image
